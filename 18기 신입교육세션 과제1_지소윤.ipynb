1.
데이터명: 프로스포츠 관람 데이터

출처: 공공데이터 포털 (https://www.data.go.kr/data/34867234/linkedData.do)

관심 이유: 각 지역별 프로스포츠 관중 및 경기 지표를 활용해서 스포츠산업 지원사업 기획을 해볼 수 있을 거 같다고 생각했다.

특징:
- 전국 광역자치단체 별 주 단위의 프로스포츠 관람객 수와 경기 건 수에 대한 데이터를 제공해준다. 
- PK와 NOT NULL 여부 정보를 제공한다.
- 비주기적으로 데이터가 업데이트 된다.

2.
데이터명: Breast Cancer Dataset

출처: kaggle (https://www.kaggle.com/datasets/mostafaabbac/breast-cancer-dataset)

관심 이유: 최근에 주변 지인분들 중 유방암 확진자가 적지 않게 생기고 있어 관심이 갔다. 이러한 데이터들을 활용해 분석하여 악성 유방암 세포의 특징을 알아보고 싶다. 

특징: 
- 유방 종양의 세침흡인 검사(FNA) 이미지를 기반으로 세포 핵의 형태적 특성을 수치화한 데이터이다. 
- 각 변수는 핵의 크기, 형태, 경계 특성 등을 나타내며, 양성/악성 종양 분류를 목적으로 한다.
- 569개의 사례가 있고 32개의 변수(ID, 진단, 30개의 실값 입력 특징)를 보여주고 있다.
