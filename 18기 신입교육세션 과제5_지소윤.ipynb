```python
data_train = data_train.drop(columns=['Support Calls_group','Payment Delay_group'])
```

```python
data_train = pd.get_dummies(data_train, columns=['Gender','Subscription Type','Contract Length'], drop_first=True)
data_test = pd.get_dummies(data_test, columns=['Gender','Subscription Type','Contract Length'], drop_first=True)
# ë¬¸ìì—´ì€ ëª¨ë¸ì´ ëª» ì½ìœ¼ë¯€ë¡œ ë²”ì£¼í˜• ë³€ìˆ˜ ì¸ì½”ë”©í•˜ê¸°
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, classification_report
```

```python
# train, test ë°ì´í„° ë‚˜ëˆ„ê¸°
Y_train = data_train['Churn']
X_train = data_train.drop(['Churn'], axis=1, inplace=False)
Y_test = data_test['Churn']
X_test = data_test.drop(['Churn'], axis=1, inplace=False)
```

```python
# train, valid ë°ì´í„° ë‚˜ëˆ„ê¸°
X_tr, X_val, y_tr, y_val = train_test_split(
    X_train, Y_train,
    test_size=0.2,
    random_state=42,
    stratify=Y_train
)
```

```python
# Logistic Regression
lr = LogisticRegression(max_iter=5000, random_state=42)
lr.fit(X_tr, y_tr)

# ê²€ì¦ í‰ê°€
val_pred = lr.predict(X_val)
val_proba = lr.predict_proba(X_val)[:, 1]

print("Accuracy:", accuracy_score(y_val, val_pred))
print("F1:", f1_score(y_val, val_pred))
print("ROC-AUC:", roc_auc_score(y_val, val_proba))
print("\n", classification_report(y_val, val_pred))
```

```python
# Gradient Boosting
gb = GradientBoostingClassifier(random_state=42)
gb.fit(X_tr, y_tr)

# ê²€ì¦ í‰ê°€
val_pred = gb.predict(X_val)
val_proba = gb.predict_proba(X_val)[:, 1]

print("Accuracy:", accuracy_score(y_val, val_pred))
print("F1:", f1_score(y_val, val_pred))
print("ROC-AUC:", roc_auc_score(y_val, val_proba))
print("\n", classification_report(y_val, val_pred))
```

```python
# Randomforest
rf = RandomForestClassifier(
    n_estimators=300,
    random_state=42,
    class_weight="balanced"
)
rf.fit(X_tr, y_tr)

# 6) ê²€ì¦ í‰ê°€
val_pred = rf.predict(X_val)
val_proba = rf.predict_proba(X_val)[:, 1]

print("Accuracy:", accuracy_score(y_val, val_pred))
print("F1:", f1_score(y_val, val_pred))
print("ROC-AUC:", roc_auc_score(y_val, val_proba))
print("\n", classification_report(y_val, val_pred))
```

* ë¡œì§€ìŠ¤í‹± íšŒê·€ëŠ” ì„±ëŠ¥ì´ ë‹¤ë¥¸ ë‘ ëª¨ë¸ì— ë¹„í•´ í˜„ì €íˆ ë‚®ê³ , ì„ í˜• ëª¨ë¸ì˜ í•œê³„ê°€ ìˆìœ¼ë¯€ë¡œ ì±„íƒí•˜ì§€ ì•Šì•˜ë‹¤.
